# SQL Assigment No.1
Top 10 city with the most order

```python
# Top 10 city with the most order
q1 = """
SELECT  s.seller_city, COUNT(oi.order_id) as order_count 
FROM sellers s
JOIN order_item oi ON s.seller_id = oi.seller_id
GROUP BY s.seller_city
ORDER BY order_count DESC
LIMIT 10
"""
ps.sqldf(q1, locals())
```

# SQL Assigment No.2
Top 10 Product Category with the most order using credit card

```python
# Top 10 Product Category with the most order using credit card
q2 = """
SELECT p.product_category_name, COUNT(oi.order_id) as order_count
FROM products p
JOIN order_item oi ON p.product_id = oi.product_id
JOIN order_payment op ON oi.order_id = op.order_id
WHERE op.payment_type = "credit_card"
GROUP BY p.product_category_name
ORDER BY order_count DESC
LIMIT 10
"""

ps.sqldf(q2, locals())
```

```python
# Payment Value = (Total Item * Price) + (Total Item * Freight Value)
order_payment.loc[order_payment['order_id'] == "00143d0f86d6fbd9f9b38ab440ac16f5"]
```

```python
# Top 10 Cities with the most highest payment value
# Payment Value = (Total Item * Price) + (Total Item * Freight Value)
q3 = """
with cte as (
SELECT order_id, seller_id, SUM(price+freight_value) as payment_value
FROM order_item
GROUP BY order_id
)

SELECT s.seller_city, 
    SUM(payment_value) as sum_payment_value, 
    s.seller_state, 
    RANK() OVER(PARTITION BY s.seller_state ORDER BY SUM(payment_value) DESC) Rank
FROM cte cte
LEFT JOIN sellers s ON cte.seller_id = s.seller_id
GROUP BY s.seller_city
ORDER BY seller_state DESC
"""

ps.sqldf(q3, locals())
```

# SQL Assigment No.5
Top 10 product category with the worst freight efficiency

```python
# Top 10 product category with the worst freight efficiency
# Assuming Freight efficiency = (freight_total/total_price)*100
q5 = """
with cte as (
SELECT order_id, product_id, ((SUM(freight_value)/SUM(price+freight_value))*100) as freight_efficiency
FROM order_item
GROUP BY order_id
)

SELECT p.product_category_name, ROUND(AVG(freight_efficiency),2) as avg_freight_efficiency
FROM cte cte
LEFT JOIN products p ON cte.product_id = p.product_id
GROUP BY p.product_category_name
ORDER BY avg_freight_efficiency DESC
LIMIT 10
"""
ps.sqldf(q5, locals())
```

# SQL Assigment No.6 [Data Preprocessing (Bonus)]
From SQL Assigment No.5, create a new column called "freight_efficiency", which categorizes the following:\
a. X <= 5% : Very Efficient\
b. 5% < x <= 15% : Efficient\
c. 15% < x <= 25% : Not Efficient\
d. <25% : Very Ineffecient

```python
# Top 10 product category with the worst freight efficiency
# Assuming Freight efficiency = (freight_value/total_price)*100
q6 = """
with cte as (
SELECT order_id, product_id, ((SUM(freight_value)/SUM(price+freight_value))*100) as freight_efficiency
FROM order_item
GROUP BY order_id
)

SELECT p.product_category_name, ROUND(AVG(freight_efficiency),2) as avg_freight_efficiency
FROM cte cte
LEFT JOIN products p ON cte.product_id = p.product_id
GROUP BY p.product_category_name
ORDER BY avg_freight_efficiency DESC
"""

df = ps.sqldf(q6, locals())
ps.sqldf(q6, locals())
```

```python
def freight_convert(freight_efficiency):
    if freight_efficiency <= 5.0:
        return "Very Efficienct"
    elif freight_efficiency > 5.0 and freight_efficiency<= 15.0:
        return "Efficient"
    elif freight_efficiency > 15.0 and freight_efficiency <= 25.0:
        return "Not Efficient"
    else:
        return "Very Inefficient"

df["freight_efficiency"] = df['avg_freight_efficiency'].apply(lambda x: freight_convert(x))
pd.set_option("display.max_rows", None, "display.max_columns", None)
df
```
